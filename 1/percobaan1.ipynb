```python
import numpy as np
from numpy import random
from scipy.fftpack import fft
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
# Signs samplying rate of 10kHz, nyq = 5000

fs = 10000
t = np.linspace(0, 1, fs+1)

sig_50Hz = np.sin(2*np.pi*50*t)
sig_110Hz = np.sin(2*np.pi*110*t)
sig_170Hz = np.sin(2*np.pi*170*t)
sig_230Hz = np.sin(2*np.pi*230*t)

sig_50Hz_110Hz_170Hz_230Hz = sig_50Hz+sig_110Hz+sig_170Hz+sig_230Hz

sig_random = 0.001*random.rand(len(t))

sig_50Hz_110Hz_170Hz_230Hz_random = sig_50Hz_110Hz_170Hz_230Hz + sig_random
```

```python
freq_domain_signal = fft(sig_50Hz_110Hz_170Hz_230Hz_random)
```

```python
style.use('dark_background')
```

```python
f,plt_arr = plt.subplots(5, sharex=True)
f.suptitle("Multifrequency and Random Signals", y=1.15)

plt_arr[0].plot(sig_50Hz_110Hz_170Hz_230Hz, color='red')
plt_arr[0].set_title("Multifrequency Signal", color='red')

plt_arr[1].plot(sig_random, color='blue')
plt_arr[1].set_title("Random Signal", color='blue')

plt_arr[2].plot(sig_50Hz_110Hz_170Hz_230Hz_random, color='yellow')
plt_arr[2].set_title("Multifrequency and Random Signals", color='yellow')

plt_arr[3].plot(freq_domain_signal, color='cyan')
plt_arr[3].set_title("Frequency Domain (FFT)", color='cyan')

magnitude = np.abs(freq_domain_signal)
plt_arr[4].plot(magnitude, color='cyan')
plt_arr[4].set_title("Magnitude Frequency Domain", color='cyan')

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=1,wspace=1,hspace=1)
plt.show()
```
